```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
%matplotlib inline
import seaborn as sns
```

```python
s1  = Series(np.random.randn(1000))
plt.hist(s1,rwidth=0.9)
```

```python
s1.plot(kind='kde')
```

```python
sns.distplot(s1,hist=True,kde=True)
```

```python
sns.distplot(s1,hist=True,kde=True,rug=True)
```

```python
sns.distplot(s1,hist=True,kde=False,bins=20)
```

```python
sns.kdeplot(s1,shade=True,color='r')
```

```python
sns.rugplot(s1)
```
