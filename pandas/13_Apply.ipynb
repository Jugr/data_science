```python
items = df0['Shrt_Desc'][0].split(' ')
# items
a = []
a += [item for item in items if item != items[0]]
```

```python
def foo(line):
    items = line.split(' ')
    a = []
    a += [item for item in items if item != items[0]]
    return Series([items[0],a])
```

```python
def foo2(line):
    items = line.split(' ')
    item0 = items[0]
    del items[0]
    return Series([item0,items])
```

```python
def foo3(line):
    items = line.split(' ')
    return Series([items[0],items[1]])
```

```python
df_tmp = df0['Shrt_Desc'].apply(foo)
```
