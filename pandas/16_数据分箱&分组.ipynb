```python
import pandas as pd
import numpy as np
from pandas import Series , DataFrame
```

```python
score_list = np.random.randint(25,100,size=20)
score_list
```

```python
bins = [0,59,70,80,100]
```

```python
score_cat = pd.cut(score_list,bins)
score_cat
```

```python
pd.value_counts(score_cat)
```

```python
df = DataFrame()
```

```python
df['score'] = score_list
df
```

```python
df['name'] = [pd.util.testing.rands(3) for i in range(20)]
```

```python
df
```

```python
df['Categories'] = pd.cut(df['score'],bins,labels=['Low','ok','good','great'])
df
```
