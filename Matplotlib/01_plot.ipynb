```python
t = np.arange(0.,2.,.1)
t.size
```

```python
s = np.sin(t*np.pi)
s.size
```

```python
plt.plot(t,s,'r--',label='a legend')
plt.plot(t*2,s,'b--',label='b legend')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('this is title')
plt.legend()
```
