```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
import pandas as pd
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind='line',grid=True,label='s1',title='this is s1')
s2.plot(kind='line',grid=True,label='s2')
plt.legend()
```

```python
fig,ax = plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
```

```python
fig,ax = plt.subplots(2,1)
s1[0:10].plot(ax=ax[0],kind='bar',grid=True,label='s1',title='this is s1')
s2.plot(ax=ax[1],grid=True,label='s2',title='this is s2')
plt.legend()
```
